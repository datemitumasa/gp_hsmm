```python
import numpy as np
import matplotlib.pyplot as plt
import glob
```

```python
CLS = "./class{0:03d}.npy"
C = len(glob.glob("./class*.npy"))
print( glob.glob("./class*.npy"))
cls = []
for i in range(C):
    p =np.load(CLS.format(i),allow_pickle=True,encoding='bytes')
    cls.append(p)
```

```python
CSV = "./GP_m{0:d}.csv"
csv =  []
for i in range(C):
    c = np.loadtxt(CSV.format(i),delimiter=",")
    csv.append(c)
```

```python
csv[0][:,0]
```

```python
base = [0.0] * 13
for i in range(C):
    plt.figure(figsize=[10,10])
    plt.plot(range(13),base,"k--")    
    plt.plot(range(13),csv[i][:,0],"r-")
    plt.plot(range(13),csv[i][:,1],"g-")
    plt.plot(range(13),csv[i][:,2],"b-")
    plt.plot(range(13),csv[i][:,3],"k-")
    plt.xlim(0,12)
    plt.ylim(-0.2,1.5)

    plt.show()
    plt.close()
```
